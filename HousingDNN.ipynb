```
import keras 
import numpy as np
import pandas as pd
```

```
df = pd.read_csv("drive/MyDrive/Colab Notebooks/DNN's/HousingPrices.csv")
df.head()
```

```
X = df.drop(columns=['SalePrice'])#all columns exept for the SalePrice column
Y = df[['SalePrice']]#just the SalePrice column
```

```
model = keras.Sequential()
#adding 3 layers of neurons
model.add(keras.layers.Dense(8, activation='relu', input_shape = (8, )))
model.add(keras.layers.Dense(8, activation='relu'))
model.add(keras.layers.Dense(1))#1 layer because is just 1 single value; SalePrice

model.compile(optimizer='adam', loss='mean_squared_error')#compile and select optimizer
```

```
#training model part
model.fit(X, Y, epochs=30, callbacks=[keras.callbacks.EarlyStopping(patience=5)])#this callback is to avoid overfitting
```

```
test_data = np.array([2003,	854,	1710,	2,	1,	3,	8,	2008])

print(model.predict(test_data.reshape(1, 8), batch_size=1))
```
